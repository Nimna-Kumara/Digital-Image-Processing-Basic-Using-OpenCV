# Introduction to Image Processing with OpenCV

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

## Read an image

### img = cv.imread(image_name, image_mode)
image_name: image name if it is in working directory or fullpath of the  image otherwise  

image_mode: is a flag which specifies the way image should be,  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• cv2.IMREAD_COLOR or 1: Loads an image in RGB mode  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• cv2.IMREAD_GARYSCALE or 0: Loads an image in grayscale mode  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• cv2.IMREAD_UNCHANGED or -1: Loads an image with alpha channel if applicable  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;• cv2.IMREAD_ANYCOLOR or 4: Loads image in any possible color format  

```python
img_color = cv.imread("images/messi.jpg",cv.IMREAD_COLOR)
img_gray = cv.imread("images/messi.jpg",cv.IMREAD_GRAYSCALE)
img_unchanged = cv.imread("images/messi.jpg",cv.IMREAD_UNCHANGED)
img_anycolor = cv.imread("images/messi.jpg",cv.IMREAD_ANYCOLOR)
```

```python
images = [img_color, img_gray, img_unchanged, img_anycolor]
titles = ["cv.IMREAD_COLOR", "cv.IMREAD_GRAYSCALE", "cv.IMREAD_UNCHANGED", "cv.IMREAD_ANYCOLO"]

for i in range(len(images)):
    plt.subplot(2,2,i+1)
    plt.title(titles[i],fontsize=8)
    plt.imshow(images[i])
    plt.axis('off')
plt.show()
```

## Display an image using OpenCV

### cv2.imshow( window_name , image_array )  

&nbsp;&nbsp;&nbsp;window_name : a name for the cv2 window  
&nbsp;&nbsp;&nbsp;image_array : opened image as an array-like object  
 
• Additional functions required  
&nbsp;&nbsp;&nbsp;cv2.waitKey() : allows users to display a window for given milliseconds or until any key is pressed  
&nbsp;&nbsp;&nbsp;cv2.destroyAllWindows(): Destroys all the windows created  

```python
img = cv.imread("images/messi.jpg", cv.IMREAD_COLOR)

cv.imshow('Main', img)
cv.waitKey(5000)
cv.destroyAllWindows()
```

## Display multiple images using OpenCV

### • Using Horizontal Stacking

```python
img = cv.imread("images/messi.jpg", cv.IMREAD_COLOR)

cv.namedWindow("main1", cv.WINDOW_NORMAL)
cv.imshow("main", np.hstack((img, img, img)))
cv.waitKey(5000)
cv.destroyAllWindows()
```

### • Using Vertical Stacking

```python
img = cv.imread("images/messi.jpg", cv.IMREAD_COLOR)

cv.namedWindow("main1", cv.WINDOW_NORMAL)
cv.imshow("main", np.vstack((img, img, img)))
cv.waitKey(5000)
cv.destroyAllWindows()
```

## Changing Color space

img_ = cv2.cvtColor( input_image, conversion_type)  
&nbsp;&nbsp;&nbsp;&nbsp;input_image : opened image as an array-like object  
&nbsp;&nbsp;&nbsp;&nbsp;conversion_type : a flag which specifies the way image should be converted  

• BGR to HSV : cv2.COLOR_BGR2HSV  
• BGR to Gray : cv2.COLOR_BGR2GRAY  
• BGR to RGB : cv2.COLOR_BGR2RGB  
• BGR to RGBA: cv2.COLOR_BGR2RGBA  

```python
img = cv.imread("images/messi.jpg", cv.IMREAD_COLOR)

img_hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_RGB = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_RGBA = cv.cvtColor(img, cv.COLOR_BGR2RGBA)
```

```python
images = [img_hsv, img_gray, img_RGB, img_RGBA]
titles = ["cv.COLOR_BGR2HSV", "cv.COLOR_BGR2GRAY", "cv.COLOR_BGR2RGB", "cv.COLOR_BGR2RGBA"]

for i in range(len(images)):
    plt.subplot(2,2,i+1)
    plt.title(titles[i],fontsize=8)
    plt.imshow(images[i])
    plt.axis('off')
plt.show()
```

## Color Regions Detection

```python
img = cv.imread("images/apple.png", cv.IMREAD_COLOR)
```

```python
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
hsv_img = cv.cvtColor(img, cv.COLOR_RGB2HSV)

# Define red color range (in HSV, red wraps around the hue scale)
lower_red1 = np.array([0, 100, 70])
upper_red1 = np.array([10, 255, 255])
lower_red2 = np.array([170, 120, 70])
upper_red2 = np.array([180, 255, 255])

# Create masks for both red ranges
mask1 = cv.inRange(hsv_img, lower_red1, upper_red1)
mask2 = cv.inRange(hsv_img, lower_red2, upper_red2)
mask = cv.bitwise_or(mask1, mask2)

# Optional: apply some morphological operations to clean the mask
kernel = np.ones((5, 5), np.uint8)
mask = cv.morphologyEx(mask, cv.MORPH_OPEN, kernel)
mask = cv.morphologyEx(mask, cv.MORPH_DILATE, kernel)

# Extract the red object
result = cv.bitwise_and(img, img, mask=mask)
```

```python
plt.subplot(1,3,1)
plt.title("Original Image", fontsize=8)
plt.imshow(img)
plt.axis('off')

plt.subplot(1,3,2)
plt.title( "Mask", fontsize=8)
plt.imshow(mask,'gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Apple Extracted", fontsize=8)
plt.imshow(result)
plt.axis('off')
plt.savefig("results/appleExtracted.png")
plt.show()
```
